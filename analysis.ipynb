```python
# ==============================
# SALES DATA ANALYSIS PROJECT
# (BASED ON YOUR ACTUAL COLUMNS)
# ==============================

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("SampleSuperstore.csv", encoding='latin1')

# ------------------------------
# Data Understanding
# ------------------------------
print("First 5 rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nColumns:")
print(df.columns)

# ------------------------------
# Data Cleaning
# ------------------------------
# Remove duplicates
df = df.drop_duplicates()

# ------------------------------
# Basic Analysis
# ------------------------------
# Total Sales
print("\nTotal Sales:", df['Sales'].sum())

# Total Profit
print("Total Profit:", df['Profit'].sum())

# Sales by Region
region_sales = df.groupby('Region')['Sales'].sum()
print("\nSales by Region:\n", region_sales)

# Sales by Category
category_sales = df.groupby('Category')['Sales'].sum()
print("\nSales by Category:\n", category_sales)

# Top 5 States
top_states = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head()
print("\nTop 5 States:\n", top_states)

# ------------------------------
# Visualization
# ------------------------------

# 1. Sales by Category
category_sales.plot(kind='bar')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()

# 2. Sales by Region
region_sales.plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()

# 3. Profit by Category
df.groupby('Category')['Profit'].sum().plot(kind='bar')
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()

# 4. Sales by State (Top 5)
top_states.plot(kind='bar')
plt.title("Top 5 States by Sales")
plt.xlabel("State")
plt.ylabel("Sales")
plt.show()


# Create Profit Margin column
df['Profit_Margin'] = (df['Profit'] / df['Sales']) * 100

# Average Profit Margin by Category
profit_margin = df.groupby('Category')['Profit_Margin'].mean()

print("\nAverage Profit Margin by Category:\n", profit_margin)

# Plot
profit_margin.plot(kind='bar')
plt.title("Average Profit Margin by Category (%)")
plt.xlabel("Category")
plt.ylabel("Profit Margin (%)")
plt.show()

print("\n✅ Analysis Completed Successfully!")
```
